# The West in Jobs

This is a rendition of the tech culture as viewed by job postings on indeed.com.  This is not meant to be scientifically precise as job postings =/= actual hiring practices.  However it should paint a picture of which cities in the west are tech centers, how those cities compare world wide, and which technologies are in demand.

#### The first thing we'll do is load up the modules necessary for the code:

```python
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from matplotlib import style
import urllib2
```

#### The second thing we do is setup the functions that will get the job numbers and plot the graphs.

```python
def find_jobs(job, location):
    job_url = "+" + str(job)
    job_url = job_url.replace("/", "%2F")
    loc_url = "+" + location
    loc_url = loc_url.replace(",", "%2C")
    url = "http://www.indeed.com/jobs?q=" + job_url + "&l=" + loc_url
    soup = BeautifulSoup(urllib2.urlopen(url).read(), "lxml")
    line = soup.find("meta", {"name":"description"})['content']
    njobs = line.split()[0].replace(",", "")
    try:
        njobs = int(njobs)
    except ValueError:
        njobs = 0
    return njobs

def plot_jobs(jobs_per_skill):
    style.use('ggplot')
    params = {'axes.labelsize': 'x-small'}
    plt.bar(range(len(jobs_per_skill)), jobs_per_skill.values(), align='center')
    plt.xticks(range(len(jobs_per_skill)), jobs_per_skill.keys(),fontsize = 7)
    plt.ylabel('Job Postings')
    plt.show(params)

    
```

#### Next, we load up some skills we are interested in analyzing.  We'll look at the demand for these skills world wide.

```python
skills = [ "aws","openshift", "vmware", "docker","openstack","ansible"]
locations = ("any")

jobs_for_skill ={}

for skill in skills:
    jobs_for_skill.update({skill: find_jobs(skill,locations)})

plot_jobs(jobs_for_skill)
```

#### Now we'll look at which western cities have the highest demand for the linux skill:

```python
skills = ["linux"]
locations = ["seattle","phoenix","los_angeles","bay_area","portland"]

jobs_for_skill ={}

for location in locations:
    jobs_for_skill.update({location: find_jobs(skills,location)})


plot_jobs(jobs_for_skill)
```
